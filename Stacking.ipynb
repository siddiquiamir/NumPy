# Stacking

```python
import numpy as np
```

```python
a = np.array([1,2,3,4,5])
print(a)
```

```python
b = np.array([6,7,8,9,10])
print(b)
```

```python
np.hstack((a,b))
```

```python
np.vstack((a,b))
```
